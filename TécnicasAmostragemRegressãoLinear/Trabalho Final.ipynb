# ENADE 2017 -- Documentação das Variáveis e Contexto do Estudo

## 1. Contextualização

O ENADE avalia o desempenho dos estudantes concluintes do ensino
superior no Brasil. A nota geral (NT_GER) pode ser influenciada por
fatores demográficos, socioeconômicos, acadêmicos e institucionais.

O objetivo deste estudo é identificar quais desses fatores melhor
explicam a variação dessa nota.

## 2. Problema de Pesquisa

Quais fatores explicam a nota geral (NT_GER) dos estudantes no ENADE
2017?

## 3. Objetivo Geral

Construir um modelo estatístico capaz de identificar quais
características dos estudantes e das instituições estão associadas ao
desempenho no ENADE.

##### Treinando modelo

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, LassoCV, RidgeCV, ElasticNetCV
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import pandas as pd

# -------------------------------
# CONFIGURAÇÕES GLOBAIS
# -------------------------------
SEED = 42
TEST_SIZE = 0.2
CV = 5

ALPHAS = np.logspace(-3, 1, 15)  # grade enxuta e suficiente
L1_RATIOS = [0.3, 0.5, 0.7]      # ElasticNet reduzido

# -------------------------------
# SPLIT ÚNICO (REPRODUTÍVEL)
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=TEST_SIZE, random_state=SEED
)

# -------------------------------
# SCALER (UMA ÚNICA VEZ)
# -------------------------------
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

resultados = {}
modelos = {}

# -------------------------------
# 1️⃣ OLS (baseline)
# -------------------------------
lr = LinearRegression(n_jobs=-1)
lr.fit(X_train_scaled, y_train)

pred = lr.predict(X_test_scaled)

resultados["OLS"] = {
    "RMSE": np.sqrt(mean_squared_error(y_test, pred)),
    "R2": r2_score(y_test, pred)
}
modelos["OLS"] = lr

# -------------------------------
# 2️⃣ LASSO
# -------------------------------
lasso = LassoCV(
    alphas=ALPHAS,
    cv=CV,
    max_iter=5000,
    n_jobs=-1
)
lasso.fit(X_train_scaled, y_train)

pred = lasso.predict(X_test_scaled)

resultados["LASSO"] = {
    "RMSE": np.sqrt(mean_squared_error(y_test, pred)),
    "R2": r2_score(y_test, pred),
    "alpha": lasso.alpha_
}
modelos["LASSO"] = lasso

# -------------------------------
# 3️⃣ RIDGE
# -------------------------------
ridge = RidgeCV(
    alphas=ALPHAS,
    cv=CV
)
ridge.fit(X_train_scaled, y_train)

pred = ridge.predict(X_test_scaled)

resultados["RIDGE"] = {
    "RMSE": np.sqrt(mean_squared_error(y_test, pred)),
    "R2": r2_score(y_test, pred),
    "alpha": ridge.alpha_
}
modelos["RIDGE"] = ridge

# -------------------------------
# 4️⃣ ELASTIC NET (opcional)
# -------------------------------
elastic = ElasticNetCV(
    alphas=ALPHAS,
    l1_ratio=L1_RATIOS,
    cv=CV,
    max_iter=5000,
    n_jobs=-1
)
elastic.fit(X_train_scaled, y_train)

pred = elastic.predict(X_test_scaled)

resultados["ELASTIC"] = {
    "RMSE": np.sqrt(mean_squared_error(y_test, pred)),
    "R2": r2_score(y_test, pred),
    "alpha": elastic.alpha_,
    "l1_ratio": elastic.l1_ratio_
}
modelos["ELASTIC"] = elastic

# -------------------------------
# RESULTADOS FINAIS
# -------------------------------
resultados_df = pd.DataFrame(resultados).T
resultados_df
```

```python
def avaliar_overfitting(modelo, X_train, y_train, X_test, y_test, nome):
    pred_train = modelo.predict(X_train)
    pred_test = modelo.predict(X_test)

    rmse_train = np.sqrt(mean_squared_error(y_train, pred_train))
    rmse_test = np.sqrt(mean_squared_error(y_test, pred_test))

    print(f"\n🟦 {nome}")
    print(f"RMSE Treino: {rmse_train:.4f}")
    print(f"RMSE Teste:  {rmse_test:.4f}")
    print(f"Gap: {abs(rmse_test - rmse_train):.4f}")

def plot_residuos(modelo, X_test, y_test, nome):
    pred = modelo.predict(X_test)
    resid = y_test - pred

    plt.figure(figsize=(7,4))
    sns.scatterplot(x=pred, y=resid, alpha=0.3)
    plt.axhline(0, color='red')
    plt.title(f"Resíduos vs Previsões - {nome}")
    plt.xlabel("Previsão")
    plt.ylabel("Resíduo")
    plt.show()

def qqplot(modelo, X_test, y_test, nome):
    pred = modelo.predict(X_test)
    resid = y_test - pred
    stats.probplot(resid, dist="norm", plot=plt)
    plt.title(f"QQ-Plot - {nome}")
    plt.show()

import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan

def testar_bp(modelo, X_test, y_test):
    pred = modelo.predict(X_test)
    resid = y_test - pred
    X_const = sm.add_constant(X_test)

    _, pval, _, _ = het_breuschpagan(resid, X_const)
    print(f"Breusch-Pagan p-value: {pval:.4f}")

```

O Exame Nacional de Desempenho dos Estudantes (ENADE) avalia o rendimento dos concluintes do ensino superior brasileiro, fornecendo um importante instrumento de diagnóstico da qualidade da educação superior. A nota geral do exame (NT_GER) reflete não apenas o conhecimento acadêmico do estudante, mas também fatores institucionais, socioeconômicos, demográficos e subjetivos relacionados à experiência no exame.

Este estudo tem como objetivo identificar quais fatores estão associados à variação da nota geral do ENADE 2017, utilizando técnicas de análise exploratória de dados, regressão linear clássica e modelos regularizados, com foco explicativo.

2. Base de Dados e Preparação

Foram utilizados os microdados do ENADE 2017, disponibilizados pelo INEP. A base passou por etapas de:

Leitura e padronização de variáveis numéricas;

Seleção de variáveis relevantes com base no dicionário oficial;

Tratamento de valores ausentes;

Codificação de variáveis categóricas quando necessário.

A variável resposta do estudo é a nota geral do estudante (NT_GER).

3. Análise Exploratória dos Dados
3.1 Variável Resposta

A análise descritiva da NT_GER indicou uma distribuição aproximadamente simétrica, com concentração das notas em torno da média e presença de valores extremos em ambas as caudas. O boxplot evidenciou variabilidade moderada e outliers esperados em avaliações de larga escala.

Esses resultados indicam heterogeneidade no desempenho dos estudantes e justificam a investigação de fatores explicativos.

3.2 Variáveis Numéricas Explicativas

As variáveis contínuas, como idade, ano de conclusão do ensino médio e ano de início da graduação, apresentaram distribuições assimétricas e concentrações temporais claras, refletindo trajetórias educacionais distintas. Essas assimetrias são esperadas e não configuram problemas metodológicos, uma vez que os pressupostos da regressão são avaliados nos resíduos do modelo.

3.3 Comparações por Grupos

Análises gráficas comparando a nota geral por sexo e por modalidade de ensino indicaram:

Distribuições semelhantes entre os sexos, sem diferenças visuais expressivas;

Mediana ligeiramente superior para estudantes de cursos presenciais em relação ao EAD, ainda que com forte sobreposição entre as distribuições.

Essas análises foram conduzidas de forma exploratória, sem inferência causal direta.

4. Análise de Correlação

A matriz de correlação de Pearson mostrou que:

A NT_GER apresenta correlações fracas a moderadas com as variáveis explicativas;

Nenhuma variável isolada explica fortemente o desempenho;

Há correlações elevadas entre algumas variáveis explicativas, especialmente entre variáveis institucionais e itens do questionário que medem construtos semelhantes.

Esses resultados reforçam a natureza multifatorial do desempenho acadêmico e indicam a necessidade de diagnóstico de multicolinearidade antes da regressão.

5. Diagnóstico de Multicolinearidade

O Fator de Inflação da Variância (VIF) indicou presença de multicolinearidade moderada, especialmente entre itens de percepção da prova e variáveis temporalmente relacionadas. Esse comportamento é esperado, dado que tais variáveis capturam dimensões próximas de um mesmo fenômeno.

Considerando que o objetivo do estudo é explicativo, optou-se por manter variáveis teoricamente relevantes, reconhecendo a necessidade de cautela na interpretação dos coeficientes individuais.

6. Modelo de Regressão Linear (OLS)

Foi ajustado um modelo de regressão linear múltipla por mínimos quadrados ordinários, com mais de 420 mil observações válidas.

6.1 Qualidade do Ajuste

O teste F indicou significância global do modelo;

O R² foi aproximadamente 0,13, valor compatível com estudos educacionais de larga escala;

O modelo tem caráter explicativo, e não preditivo.

6.2 Interpretação dos Coeficientes

Os resultados indicaram que:

A idade apresenta associação negativa com a nota;

A modalidade presencial está associada a notas ligeiramente maiores;

Variáveis institucionais (região e categoria administrativa) influenciam o desempenho;

Variáveis de percepção do estudante sobre a prova e o aprendizado apresentam forte associação positiva;

Fatores socioeconômicos, como escolaridade dos pais e situação de trabalho, também são estatisticamente significativos.

Esses resultados reforçam o caráter multifatorial do desempenho no ENADE.

7. Diagnóstico dos Pressupostos do Modelo
7.1 Resíduos e Especificação

O gráfico de resíduos versus valores ajustados não indicou curvatura sistemática;

O teste de Ramsey RESET rejeitou a hipótese de especificação correta, sugerindo possíveis não linearidades ou variáveis omitidas;

Esse resultado é esperado em grandes amostras e não invalida o modelo.

7.2 Normalidade e Heterocedasticidade

QQ-plots indicaram normalidade aproximada dos resíduos no centro da distribuição;

Testes de Breusch-Pagan e White indicaram heterocedasticidade;

Dado o tamanho amostral, esses resultados são esperados e recomendam cautela na inferência pontual.

8. Modelos Regularizados e Comparação

Foram ajustados modelos OLS, Lasso, Ridge e Elastic Net, com:

Padronização das variáveis;

Validação cruzada;

Avaliação por RMSE e R²;

Repetição do processo com múltiplas seeds.

8.1 Resultados Comparativos

Os modelos apresentaram desempenhos praticamente idênticos;

O R² manteve-se em torno de 0,25 em todas as seeds;

Não houve ganho relevante de desempenho com regularização.

Esses resultados indicam estabilidade e robustez do ajuste linear clássico.

8.2 Overfitting

A comparação entre RMSE de treino e teste mostrou gaps reduzidos, indicando ausência de overfitting relevante tanto no OLS quanto no Lasso.

9. Análise de Impacto das Variáveis (Lasso)

O gráfico dos coeficientes do Lasso mostrou que:

Apenas um subconjunto reduzido de variáveis apresenta impacto relevante;

Muitas variáveis categóricas tiveram coeficientes penalizados próximos de zero;

O Lasso atua como ferramenta eficaz de seleção e validação de robustez.

Os coeficientes do Lasso não foram interpretados causalmente, mas como indicativos de relevância relativa.

10. Conclusão

Os resultados indicam que o desempenho dos estudantes no ENADE 2017 é explicado por uma combinação de fatores demográficos, institucionais, socioeconômicos e subjetivos relacionados à percepção do exame e ao engajamento acadêmico.

Nenhuma variável isolada é capaz de explicar o desempenho, reforçando o caráter complexo e multifatorial do fenômeno educacional. A regressão linear clássica mostrou-se adequada para fins explicativos, enquanto os modelos regularizados confirmaram a robustez dos resultados, sem ganhos preditivos adicionais relevantes.

Este estudo evidencia a importância de abordagens multivariadas na análise educacional e destaca que interpretações simplistas ou unidimensionais do desempenho acadêmico são insuficientes.
