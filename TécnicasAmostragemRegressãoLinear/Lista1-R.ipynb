#### Exercicio 1

```r
n = 25
media = 51.3
desvio_padrao = 2
ic = 1-0.95

zc = qnorm(p = 1 - ic/2, mean = 0, sd = 1)
erro_margem = zc * (desvio_padrao / sqrt(n))
intervalo_confianca = c(media - erro_margem, media + erro_margem)
print(paste("Intervalo de confiança para a média com 95% de confiança:", intervalo_confianca[1], "a", intervalo_confianca[2]))
```

#### Exercicio 3

#### Exercicio 4

```r
amostra = c(19.8, 18.5, 17.6, 16.7, 15.8, 15.4, 14.1, 13.6, 11.9, 11.4, 11.4, 8.8, 7.5, 15.4, 15.4, 19.5, 14.9, 12.7, 11.9, 11.4, 10.1, 7.9)
media = mean(amostra)
desvio_padrao = sd(amostra)
n = length(amostra)
ic = 1 - 0.95

tc = qt(p = 1 - ic/2, df = n - 1)
erro_margem = tc * (desvio_padrao / sqrt(n))
intervalo_confianca = c(media - erro_margem, media + erro_margem)
print(paste("Intervalo de confiança para a média com 95% de confiança:", intervalo_confianca[1], "a", intervalo_confianca[2]))
```

#### Exercicio 5

```r
n = 6
amostra = c(16.8, 17.2, 17.4, 16.9, 16.5, 17.1)
media = mean(amostra)
soma_quadrados = sum((amostra - media)^2)
variancia_amostra = soma_quadrados / (n - 1)
ic = 1 - 0.99
tc = qt(p = 1 - ic/2, df = n - 1)
erro_margem = tc * sqrt(variancia_amostra / n)
intervalo_confianca = c(media - erro_margem, media + erro_margem)
print(paste("Intervalo de confiança para a média com 99% de confiança:", intervalo_confianca[1], "a", intervalo_confianca[2]))
```

#### Exercicio 7

```r
p_conservador = 0.5
p_otimista = 0.8
n = 85
x = 10
p = x / n
ic = 1 - 0.95
zc = qnorm(p = 1 - ic/2, mean = 0, sd = 1)

erro_margem = zc * ((p * (1 - p) / n) ** 0.5)
intervalo_confianca = c(p - erro_margem, p + erro_margem)
print(paste("Intervalo de confiança para a proporção com 95% de confiança:", intervalo_confianca[1], "a", intervalo_confianca[2]))

erro_margem_otimista = zc * ((p_otimista * (1 - p_otimista) / n) ** 0.5)
intervalo_confianca_otimista = c(p - erro_margem_otimista, p + erro_margem_otimista)
print(paste("Intervalo de confiança para a proporção (p otimista) com 95% de confiança:", intervalo_confianca_otimista[1], "a", intervalo_confianca_otimista[2]))

erro_margem_conservador = zc * ((p_conservador * (1 - p_conservador) / n) ** 0.5)
intervalo_confianca_conservador = c(p - erro_margem_conservador, p + erro_margem_conservador)
print(paste("Intervalo de confiança para a proporção (p conservador) com 95% de confiança:", intervalo_confianca_conservador[1], "a", intervalo_confianca_conservador[2]))
```

#### Exercicio 9

```r
ic = 1 - 0.92
p = 0.1
erro = 0.02

p_conservador = 0.5
p_otimista = 0.8

zc = qnorm(p = 1 - ic/2, mean = 0, sd = 1)
n_conservador = (zc^2 * p_conservador * (1 - p_conservador)) / (erro^2)
n_otimista = (zc^2 * p_otimista * (1 - p_otimista)) / (erro^2)
n = (zc^2 * p * (1 - p)) / (erro^2)
print(paste("Tamanho da amostra necessário para um erro máximo de 0.02 com 92% de confiança:", ceiling(n)))
print(paste("Tamanho da amostra necessário (p conservador) para um erro máximo de 0.02 com 92% de confiança:", ceiling(n_conservador)))
print(paste("Tamanho da amostra necessário (p otimista) para um erro máximo de 0.02 com 92% de confiança:", ceiling(n_otimista)))
```
