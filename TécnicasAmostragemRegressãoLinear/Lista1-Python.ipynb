```python
import scipy.stats as stats
import numpy as np
```

#### Exercicio 1 

```python
n = 25
media = 51.3
desvio_padrao = 2
ic = 0.95

valor_critico = stats.norm.ppf(1 - (1 - ic) / 2)
erro_margem = valor_critico * desvio_padrao / (n ** 0.5)
intervalo_confianca = (media - erro_margem, media + erro_margem)
print("Intervalo de confiança para a média com 95% de confiança:", intervalo_confianca)
```

#### Exercicio 3

```python
e = 1.5
i = 0.995
desvio_padrao_quadrado = 23
z = stats.norm.ppf(1 - (1 - i) / 2)
desvio_padrao = desvio_padrao_quadrado ** 0.5
n = (z * desvio_padrao / e) ** 2
print("Tamanho da amostra necessário para um erro máximo de 1.5 com 99.5% de confiança:", round(n))
```

#### Exercicio 4

```python
amostra = [19.8, 18.5, 17.6, 16.7, 15.8, 15.4, 14.1, 13.6, 11.9, 11.4, 11.4, 8.8, 7.5, 15.4, 15.4, 19.5, 14.9, 12.7, 11.9, 11.4, 10.1, 7.9]
media = sum(amostra) / len(amostra)
desvio_padrao = np.std(amostra, ddof=1)


valor_critico = stats.t.ppf(1 - (1 - ic) / 2, n-1)
erro_margem = valor_critico * (desvio_padrao / (n ** 0.5))
intervalo_confianca = (media - erro_margem, media + erro_margem)
print("Intervalo de confiança para a média com 95% de confiança:", intervalo_confianca)
```

#### Exercicio 5

```python
n = 6
amostra = [16.8, 17.2, 17.4, 16.9, 16.5, 17.1]
media_amostra = sum(amostra) / n
soma_quadrados = sum((x - media_amostra) ** 2 for x in amostra)
s = (soma_quadrados / (n - 1)) ** 0.5
ic = 0.99
valor_critico = stats.t.ppf(1 - (1 - ic) / 2, n-1)
erro_margem = valor_critico * (s / (n ** 0.5))
intervalo_confianca = (media_amostra - erro_margem, media_amostra + erro_margem)
print("Intervalo de confiança para a média com 99% de confiança:", intervalo_confianca)
```

#### Exercicio 7

```python
p_conservador = 0.5
p_otimista = 0.8
n = 85
x = 10
p = x/n
ic = 0.95

valor_critico = stats.norm.ppf(1 - (1 - ic) / 2)

erro_margem = valor_critico * ((p * (1 - p)) / n) ** 0.5
intervalo_confianca = (p - erro_margem, p + erro_margem)
print("Intervalo de confiança para a proporção com 95% de confiança:", intervalo_confianca)

erro_margem_otimista = valor_critico * ((p_otimista * (1 - p_otimista)) / n) ** 0.5
intervalo_confianca = (p - erro_margem_otimista, p + erro_margem_otimista)
print("Intervalo de confiança para a proporção (p otimista) com 95% de confiança:", intervalo_confianca)

erro_margem_conservador = valor_critico * ((p_conservador * (1 - p_conservador)) / n) ** 0.5
intervalo_confianca = (p - erro_margem_conservador, p + erro_margem_conservador)
print("Intervalo de confiança para a proporção (p conservador) com 95% de confiança:", intervalo_confianca)
```

#### Exercicio 9

```python
ic = 0.92
p = 0.1
erro = 0.02

p_conservador = 0.5
p_otimista = 0.8

z = stats.norm.ppf(1 - (1 - ic) / 2)
n = ((z / erro) ** 2) * (p * (1 - p))
print("Tamanho da amostra necessário para um erro máximo de 0.02 com 92% de confiança:", round(n))

n_conservador = ((z / erro) ** 2) * (p_conservador * (1 - p_conservador))
print("Tamanho da amostra necessário para um erro máximo de 0.02 com 92% de confiança (p conservador):", round(n_conservador))

n_otimista = ((z / erro) ** 2) * (p_otimista * (1 - p_otimista))
print("Tamanho da amostra necessário para um erro máximo de 0.02 com 92% de confiança (p otimista):", round(n_otimista))

```
