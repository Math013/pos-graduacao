# ENADE 2017 -- Documentação das Variáveis e Contexto do Estudo

## 1. Contextualização

O ENADE avalia o desempenho dos estudantes concluintes do ensino
superior no Brasil. A nota geral (NT_GER) pode ser influenciada por
fatores demográficos, socioeconômicos, acadêmicos e institucionais.

O objetivo deste estudo é identificar quais desses fatores melhor
explicam a variação dessa nota.

## 2. Problema de Pesquisa

Quais fatores explicam a nota geral (NT_GER) dos estudantes no ENADE
2017?

## 3. Objetivo Geral

Construir um modelo estatístico capaz de identificar quais
características dos estudantes e das instituições estão associadas ao
desempenho no ENADE.

## 4. Objetivos Específicos

-   Processar e limpar os microdados.
-   Explorar estatisticamente as variáveis.
-   Avaliar correlações entre fatores relevantes.
-   Ajustar modelos de regressão.
-   Identificar quais fatores têm maior impacto na NT_GER.

## 5. Hipóteses

**Hipótese Nula (H0):** As variáveis analisadas não explicam a NT_GER.\
**Hipótese Alternativa (H1):** Pelo menos uma variável explica a NT_GER.

### Exemplos de hipóteses específicas

-   Idade influencia a nota?
-   Há diferença entre sexos?
-   Modalidade do curso afeta NT_GER?
-   Renda familiar impacta o desempenho?
-   Escolaridade dos pais influencia o resultado?
-   A percepção do estudante sobre a prova (CO_RS_I1--I9) se relaciona
    com NT_GER?

```python
df["NT_GER"] = pd.to_numeric(df["NT_GER"], errors="coerce")
df["TP_SEXO"] = df["TP_SEXO"].map({"M": 1, "F": 0})

map_I1 = {"A":1, "B":2, "C":3, "D":4, "E":5}
map_I4 = map_I5 = map_I6 = map_I1
map_I8 = map_I9 = map_I1

map_I3 = {"E":1, "D":2, "C":3, "B":4, "A":5}

ordinal_maps = {
    "CO_RS_I1": map_I1,
    "CO_RS_I3": map_I3,
    "CO_RS_I4": map_I4,
    "CO_RS_I5": map_I5,
    "CO_RS_I6": map_I6,
    "CO_RS_I8": map_I8,
    "CO_RS_I9": map_I9
}

for col, mapa in ordinal_maps.items():
    df[col] = df[col].map(mapa)

df = pd.get_dummies(df, columns=["CO_RS_I7"], drop_first=True)

map_escolaridade = {"A":1, "B":2, "C":3, "D":4, "E":5, "F":6}
map_renda = {"A":1, "B":2, "C":3, "D":4, "E":5, "F":6, "G":7}
map_trabalho = {"A":1, "B":2, "C":3, "D":4, "E":5}

df["QE_I04"] = df["QE_I04"].map(map_escolaridade)
df["QE_I05"] = df["QE_I05"].map(map_escolaridade)
df["QE_I08"] = df["QE_I08"].map(map_renda)
df["QE_I10"] = df["QE_I10"].map(map_trabalho)

# -----------------------------
# 6. Conferir dtype final
# -----------------------------
df.dtypes
```

##### Treinando modelo

```python
from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, LassoCV, RidgeCV, ElasticNetCV
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

def rodar_modelos_cv(seed, k=5):
    # 1) SPLIT TREINO / TESTE
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, random_state=seed
    )
    
    # 2) SCALER
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    resultados = {}
    modelos = {}

    # 3) REGRESSÃO LINEAR
    lr = LinearRegression()
    lr.fit(X_train_scaled, y_train)
    pred_lr = lr.predict(X_test_scaled)

    resultados["OLS"] = {
        "RMSE": np.sqrt(mean_squared_error(y_test, pred_lr)),
        "R2": r2_score(y_test, pred_lr)
    }
    modelos["OLS"] = lr

    # 4) LASSO
    lasso = LassoCV(
        alphas=np.logspace(-4, 1, 50),
        cv=k,
        max_iter=8000,
        n_jobs=-1
    )
    lasso.fit(X_train_scaled, y_train)
    pred_lasso = lasso.predict(X_test_scaled)

    resultados["LASSO"] = {
        "RMSE": np.sqrt(mean_squared_error(y_test, pred_lasso)),
        "R2": r2_score(y_test, pred_lasso),
        "alpha": lasso.alpha_
    }
    modelos["LASSO"] = lasso

    # 5) RIDGE
    ridge = RidgeCV(
        alphas=np.logspace(-3, 3, 50),
        cv=k
    )
    ridge.fit(X_train_scaled, y_train)
    pred_ridge = ridge.predict(X_test_scaled)

    resultados["RIDGE"] = {
        "RMSE": np.sqrt(mean_squared_error(y_test, pred_ridge)),
        "R2": r2_score(y_test, pred_ridge),
        "alpha": ridge.alpha_
    }
    modelos["RIDGE"] = ridge

    # 6) ELASTIC NET
    en = ElasticNetCV(
        l1_ratio=[0.1, 0.3, 0.5, 0.7, 0.9],
        alphas=np.logspace(-4, 1, 50),
        cv=k,
        max_iter=8000,
        n_jobs=-1
    )
    en.fit(X_train_scaled, y_train)
    pred_en = en.predict(X_test_scaled)

    resultados["ELASTIC"] = {
        "RMSE": np.sqrt(mean_squared_error(y_test, pred_en)),
        "R2": r2_score(y_test, pred_en),
        "alpha": en.alpha_,
        "l1_ratio": en.l1_ratio_
    }
    modelos["ELASTIC"] = en

    return resultados, modelos, scaler
```

```python
seeds = [13, 42, 123, 222, 999]
todos_resultados = {}

for s in seeds:
    print(f"Rodando seed {s}...")
    res, mods, _ = rodar_modelos_cv(s)
    todos_resultados[s] = res

todos_resultados_df = pd.DataFrame(todos_resultados).T
todos_resultados_df
```

```python
def avaliar_overfitting(modelo, X_train, y_train, X_test, y_test, nome):
    pred_train = modelo.predict(X_train)
    pred_test = modelo.predict(X_test)

    rmse_train = np.sqrt(mean_squared_error(y_train, pred_train))
    rmse_test = np.sqrt(mean_squared_error(y_test, pred_test))

    print(f"\n🟦 {nome}")
    print(f"RMSE Treino: {rmse_train:.4f}")
    print(f"RMSE Teste:  {rmse_test:.4f}")
    print(f"Gap: {abs(rmse_test - rmse_train):.4f}")

def plot_residuos(modelo, X_test, y_test, nome):
    pred = modelo.predict(X_test)
    resid = y_test - pred

    plt.figure(figsize=(7,4))
    sns.scatterplot(x=pred, y=resid, alpha=0.3)
    plt.axhline(0, color='red')
    plt.title(f"Resíduos vs Previsões - {nome}")
    plt.xlabel("Previsão")
    plt.ylabel("Resíduo")
    plt.show()

def qqplot(modelo, X_test, y_test, nome):
    pred = modelo.predict(X_test)
    resid = y_test - pred
    stats.probplot(resid, dist="norm", plot=plt)
    plt.title(f"QQ-Plot - {nome}")
    plt.show()

import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan

def testar_bp(modelo, X_test, y_test):
    pred = modelo.predict(X_test)
    resid = y_test - pred
    X_const = sm.add_constant(X_test)

    _, pval, _, _ = het_breuschpagan(resid, X_const)
    print(f"Breusch-Pagan p-value: {pval:.4f}")

```

3. Modelos ajustados
Foram ajustados quatro modelos de regressão linear múltipla:

3) Ridge (L2) – com seleção automática de alpha via RidgeCV (k=5).

Entre os modelos, o Elastic Net apresentou desempenho ligeiramente superior e mais estável entre seeds, sendo escolhido como modelo final para interpretação.

5. Validação de pressupostos

5.1 Overfitting / Underfitting
Para o modelo Elastic Net (seed de validação, por exemplo 42), obteve-se:

- RMSE Treino ≈ 12.1865
- RMSE Teste  ≈ 12.1867
- Gap ≈ 0.0002

A proximidade entre os erros de treino e teste indica forte capacidade de generalização: não há evidência de overfitting nem de underfitting. A regularização está funcionando de forma adequada.

5.2 Resíduos vs previsões

5.3 Normalidade dos resíduos (QQ-plot)

5.4 Heterocedasticidade (Breusch-Pagan)

6. Comparação entre OLS, LASSO, Ridge e Elastic Net

- OLS: serve como baseline, sem penalização. Apresenta desempenho semelhante em termos de RMSE e R², porém mais sensível à multicolinearidade e à presença de muitas dummies.
- LASSO: realiza seleção de variáveis, zerando alguns coeficientes. Ajuda na parcimônia do modelo, reduzindo dimensionalidade e facilitando a interpretação.
- Ridge: penaliza o tamanho dos coeficientes, mas não os zera. É particularmente útil para estabilizar estimativas na presença de alta correlação entre preditores.
- Elastic Net: combina L1 e L2, herdando vantagens de ambos. No estudo, apresentou o melhor equilíbrio entre desempenho preditivo, estabilidade e parcimônia, sendo adotado como modelo final.
