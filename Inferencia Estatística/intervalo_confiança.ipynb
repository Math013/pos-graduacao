```python
import scipy.stats as st
import numpy as np
```

```python
amostra = [501.58, 505.3, 497.62, 511.23, 502.13, 498.45, 496.38, 498.44, 498.45, 488.0,
508.1, 493.79, 505.24, 492.62, 518.15, 488.77, 495.19, 510.56, 477.39, 508.69,
488.84, 503.39, 499.12, 511.97, 484.84]

media_amostral = np.mean(amostra)
desvio_populacional = np.sqrt(100)
n = len(amostra)
nivel_confianca = 0.95

print("O tamanho da amostra é:", n)
print("A média amostral é:", media_amostral)

IC_media = st.norm.interval(nivel_confianca, loc=media_amostral, scale=desvio_populacional/np.sqrt(n))

print(f"O intervalo de {nivel_confianca*100}% de confiança para a quantidade média de gramas nos pacotes é de: ", [round(num, 2) for num in IC_media])
```

```python
amostra_criancas = [1.71, 1.06, 1.68, 2.12, 2.17, 2.71, 1.02, 2.62, 2.15, 2.22, 2.1, 1.97, 1.44, 2.48, 1.45,
2.3, 1.73, 1.0, 2.36, 0.8]

media_amostral = np.mean(amostra_criancas)
desvio_amostral = np.std(amostra_criancas, ddof=1)
n = len(amostra_criancas)

print('O tamanho da amostra é:', n)
print('A média amostral é:', round(media_amostral))
print('o desvio padrão amostral é:', round(desvio_amostral, 1))

nivel_confianca = 0.95
IC_media_95 = st.t.interval(nivel_confianca, df=n-1, loc=media_amostral, scale=desvio_amostral/np.sqrt(n))

print(f'O intervalo de {nivel_confianca*100}% de confiança para a idade média ao falar das crianças é de: ', [round(num, 2) for num in IC_media_95])

amplitude_IC = IC_media_95[1] - IC_media_95[0]
print('A amplitude do intervalo de confiança é de:', round(amplitude_IC, 2))

nivel_confianca = 0.90
IC_media_90 = st.t.interval(nivel_confianca, df=n-1, loc=media_amostral, scale=desvio_amostral/np.sqrt(n))

print(f'O intervalo de {nivel_confianca*100}% de confiança para a idade média ao falar das crianças é de: ', [round(num, 2) for num in IC_media_90])
```

```python
st.t.interval(nivel_confianca, df = n-1, loc=np.mean(amostra_criancas), scale=st.sem(amostra_criancas))
```

```python
from statsmodels.stats.proportion import proportion_confint   

n = 1500
entrevistados_sim = 1050 # Número de sucessos
p_chapeu = entrevistados_sim/n
nivel_confianca = 0.95

print("O tamanho da amostra é de:", n, " clientes.")
print("A proporção de entrevistados que acreditam no aquecimento global é de: ", p_chapeu, "\n")

print('Intervalo de confiança otimista')
ICp_95_otimista = proportion_confint(entrevistados_sim, n, alpha=1-nivel_confianca)
print("O intervalo de ", nivel_confianca*100 ,
      "% confiança para a proporção de entrevistados que acreditam no aquecimento global é de: ", [round(num,3) for num in ICp_95_otimista],"\n")

# Intervalo de confiança Conservador usando a função par 'Beta'
print('Intervalo de confiança Conservador - Beta ')
ICp_95_conservador = proportion_confint(entrevistados_sim, n, alpha=1-nivel_confianca, method = 'beta')
print("O intervalo de ", nivel_confianca*100 ,
      "% confiança para a proporção de entrevistados que acreditam no aquecimento global é de: ",[round(num,3) for num in ICp_95_conservador],"\n")
```
