```python
import scipy.stats as stats
```

```python
amostra = 50
pontuacao = 560
desvio_padrao = 25
confianca_90 = 1 - 0.90
confianca_95 = 1 - 0.95
confianca_99 = 1 - 0.99
```

```python
zc_90 = stats.norm.ppf(1 - confianca_90 / 2)
zc_95 = stats.norm.ppf(1 - confianca_95 / 2)
zc_99 = stats.norm.ppf(1 - confianca_99 / 2)

erro_90 = zc_90 * desvio_padrao / (amostra ** 0.5)
erro_95 = zc_95 * desvio_padrao / (amostra ** 0.5)
erro_99 = zc_99 * desvio_padrao / (amostra ** 0.5)
```

```python
intervalo_90 = (pontuacao - erro_90, pontuacao + erro_90)
intervalo_95 = (pontuacao - erro_95, pontuacao + erro_95)
intervalo_99 = (pontuacao - erro_99, pontuacao + erro_99)

print("Intervalo de confiança 90%:", intervalo_90)
print("Intervalo de confiança 95%:", intervalo_95)
print("Intervalo de confiança 99%:", intervalo_99)
```

```python
amplitude_90 = intervalo_90[1] - intervalo_90[0]
amplitude_95 = intervalo_95[1] - intervalo_95[0]
amplitude_99 = intervalo_99[1] - intervalo_99[0]

print("Amplitude do intervalo de confiança 90%:", amplitude_90)
print("Amplitude do intervalo de confiança 95%:", amplitude_95)
print("Amplitude do intervalo de confiança 99%:", amplitude_99)
```

```python
e_90 = 2 * erro_90
e_95 = 2 * erro_95
e_99 = 2 * erro_99

print("Erro máximo tolerado para 90% de confiança:", e_90)
print("Erro máximo tolerado para 95% de confiança:", e_95)
print("Erro máximo tolerado para 99% de confiança:", e_99)
```

```python
n = 30
x = 10
p = x / n
q = 1 - p
zc_95 = stats.norm.ppf(1 - 0.05 / 2)
erro_margem = zc_95 * ((p * q) / n) ** 0.5
intervalo_confianca = (p - erro_margem, p + erro_margem)
print("Intervalo de confiança para a proporção com 95% de confiança:", intervalo_confianca)
```

```python
lista_amostra = [661, 595 ,548, 730 ,791, 678, 672 ,491, 492, 583, 762, 624, 769, 729,734,706]
media_amostra = sum(lista_amostra) / len(lista_amostra)
desvio_padrao_amostra = (sum((x - media_amostra) ** 2 for x in lista_amostra) / (len(lista_amostra) - 1)) ** 0.5
valor_critico = stats.t.ppf(1 - 0.05 / 2, df=len(lista_amostra) - 1)
erro_margem_t = valor_critico * desvio_padrao_amostra / (len(lista_amostra) ** 0.5)
intervalo_confianca_t = (media_amostra - erro_margem_t, media_amostra + erro_margem_t)
print("Intervalo de confiança para a média com 95% de confiança (t-student):", intervalo_confianca_t)
```
