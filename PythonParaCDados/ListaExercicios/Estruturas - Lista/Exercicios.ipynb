#### Exercício 1

```python
from random import randint, sample
```

```python
lista_numeros = [randint(1, 20) for _ in range(10)]
print('Lista de numeros: ', lista_numeros)
print('Maior valor: ', max(lista_numeros))
print('Menor valor: ', min(lista_numeros))
print('Soma dos valores: ', sum(lista_numeros))
print('Média dos valores: ', sum(lista_numeros) / len(lista_numeros))
print('Números em ordem crescente: ', sorted(lista_numeros))
print('Números em ordem decrescente: ', sorted(lista_numeros, reverse=True))

valor = int(input('Digite um valor para verificar se está na lista: '))
if valor in lista_numeros:
    print(f'O valor {valor} está na lista, e aparece {lista_numeros.count(valor)} vez(es).')
else:
    print(f'O valor {valor} não está na lista.')
```

#### Exercício 2

```python
lista1 = [randint(1, 20) for _ in range(10)]
lista2 = [randint(1, 20) for _ in range(10)]

lista_soma = [a + b for a, b in zip(lista1, lista2)]
lista_multiplicacao = [a * b for a, b in zip(lista1, lista2)]

print('Lista 1: ', lista1)
print('Lista 2: ', lista2)  
print('Soma das listas: ', lista_soma)
print('Multiplicação das listas: ', lista_multiplicacao)
```

#### Exercicio 4

```python
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho',
         'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']

data = str(input('Digite uma data no formato DD/MM/AAAA: '))
dia, mes, ano = data.split('/')

data_formatada = f'{dia} de {meses[int(mes)-1]} de {ano}'
print(f'{data} - {data_formatada}')
```

#### Exercicio 5

```python
mega_sena = sample(range(1, 61), 6)
jogo = []

while len(jogo) < 6:
    numero = int(input('Digite um número entre 1 e 60: '))
    if numero < 1 or numero > 60:
        print('Número inválido! Tente novamente.')
    elif numero in jogo:
        print('Número já escolhido! Tente novamente.')
    else:
        jogo.append(numero)
        
print('Seu jogo: ', sorted(jogo))
print('Números da Mega-Sena: ', sorted(mega_sena))
```
