```python
import pandas as pd
```

```python
jogadores = pd.read_csv('jogadores.csv', sep=';')
jogadores
```

```python
jogadores = jogadores.dropna()
jogadores
```

```python
jogadores['salario_anual'] = jogadores['salario'] * 12
jogadores
```

```python
jogadores['jogador'] = jogadores['jogador'].str.upper()
jogadores
```

```python
jogadores['salario'].describe()
```

```python
jogadores['salario_anual'].describe()
```

```python
jogadores_salarios = jogadores.loc[
    jogadores['salario'] > 500000, 
    ['jogador', 'time']
]
jogadores_salarios
```

```python
jogadores_com_u = jogadores.loc[
    jogadores['jogador'].str.contains('U'), 
    ['jogador', 'time']
]
jogadores_com_u
```

```python
jogadores_ordem_decrescente = jogadores.sort_values(
    by='salario',
    ascending=False
)[['jogador', 'salario', 'time']]
jogadores_ordem_decrescente
```

```python
jogadores_ordenados = jogadores.sort_values(
    by=['time', 'salario'],
    ascending=[True, False]
)[['jogador', 'salario', 'time']]
jogadores_ordenados
```

```python
quantidade_jogadores_por_time = jogadores['time'].value_counts()
quantidade_jogadores_por_time
```

```python
media_por_time = jogadores.groupby('time')['salario'].mean()
media_por_time
```

```python
jogadores.to_csv('novo_time.csv', sep=';', index=False)
```

```python
jogadores_por_posicao = jogadores.groupby('posicao')['jogador'].count()
jogadores_por_posicao.plot(kind='bar', title='Quantidade de Jogadores por Posição')
```
