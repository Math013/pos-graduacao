#### Exercicio 2

```python
# Dicion√°rio com nome e e-mail de 5 pessoas
pessoas = {
    "Ana": "ana.silva@gmail.com",
    "Bruno": "bruno.souza@yahoo.com",
    "Carla": "carla.mendes@hotmail.com",
    "Diego": "diego.lima@outlook.com",
    "Eduarda": "eduarda.pereira@gmail.com"
}

# Consulta por nome
nome = input("Digite o nome da pessoa: ")

if nome in pessoas:
    print(f"O e-mail de {nome} √© {pessoas[nome]}")
else:
    print("Nome n√£o encontrado.")

# Consulta por e-mail
email = input("\nDigite o e-mail da pessoa: ")

encontrado = False
for n, e in pessoas.items():
    if e == email:
        print(f"O e-mail {email} pertence a {n}")
        encontrado = True
        break

if not encontrado:
    print("E-mail n√£o encontrado.")
```

#### Exercicio 6

```python
ceps = []
for i in range(5):
    cep = input(f"Digite o {i+1}¬∫ CEP (somente n√∫meros): ").strip()
    ceps.append(cep)

print("\nResultados das consultas:\n")

for cep in ceps:
    # URL da API ViaCEP
    url = f"https://viacep.com.br/ws/{cep}/json/"

    try:
        resposta = requests.get(url)
        resposta.raise_for_status()  # verifica se a requisi√ß√£o teve sucesso
        dados = resposta.json()

        # 3Ô∏è‚É£ Verifica se o CEP √© v√°lido
        if "erro" in dados:
            print(f"CEP inv√°lido: {cep}\n")
        else:
            print(f"CEP: {dados.get('cep', 'N/A')}")
            print(f"Logradouro: {dados.get('logradouro', 'N/A')}")
            print(f"Complemento: {dados.get('complemento', 'N/A')}")
            print(f"Bairro: {dados.get('bairro', 'N/A')}")
            print(f"Localidade: {dados.get('localidade', 'N/A')}")
            print(f"UF: {dados.get('uf', 'N/A')}\n")

    except requests.exceptions.RequestException as e:
        print(f"Erro ao consultar o CEP {cep}: {e}\n")
```

#### Exercicio 7

```python
from datetime import datetime

codigo = input("Digite o c√≥digo da moeda (ex: USD, EUR, BTC): ").strip().upper()
quantidade = input("Quantas cota√ß√µes deseja ver? ").strip()

url = f"https://economia.awesomeapi.com.br/json/daily/{codigo}-BRL/{quantidade}"

try:
    resposta = requests.get(url)

    dados = resposta.json()

    if not dados:
        print("Nenhum dado encontrado. Verifique o c√≥digo da moeda.")
    else:
        print(f"\n√öltimas {quantidade} cota√ß√µes de {codigo}/BRL:\n")

        # 6Ô∏è‚É£ Percorre as cota√ß√µes retornadas
        for cotacao in dados:
            timestamp = int(cotacao["timestamp"])
            data = datetime.fromtimestamp(timestamp).strftime("%d/%m/%Y %H:%M:%S")

            print(f"Data: {data}")
            print(f"Valor de venda (bid): R$ {cotacao['bid']}")
            print(f"Valor de compra (ask): R$ {cotacao['ask']}\n")

except requests.exceptions.RequestException as erro:
    print("Erro ao acessar a API:", erro)
```
