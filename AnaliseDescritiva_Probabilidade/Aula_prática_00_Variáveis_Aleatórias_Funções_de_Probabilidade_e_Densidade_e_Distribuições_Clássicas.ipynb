# Aula: Variáveis Aleatórias, Funções de Probabilidade e Densidade, e Distribuições Clássicas

### 1. Introdução às Variáveis Aleatórias
- **Variável Aleatória Discreta:** Assume valores isolados, enumeráveis (ex: número de caras em lançamentos de moedas).
- **Variável Aleatória Contínua:** Assume valor em um intervalo contínuo (ex: altura da pessoa, tempo).

### 2. Função de Probabilidade e Função Densidade
- **Função de Probabilidade (pmf) para variável discreta:** $P(X = x_i)$, soma dos valores = 1.
- **Função Densidade (pdf) para variável contínua:** $f(x)$, a probabilidade que $X$ esteja em um intervalo $[a,b]$ é a área sob a curva entre $a$ e $b$:
  $$
  P(a \leq X \leq b) = \int_a^b f(x) \, dx
  $$

### 3. Distribuições Clássicas
- **Bernoulli:** 0 ou 1 (sucesso ou fracasso). Parâmetro $p$ = probabilidade de sucesso.
  $P(X = 1) = p$, $P(X = 0) = 1 - p$.
- **Binomial:** Número de sucessos em $n$ tentativas independentes de Bernoulli. Parâmetros: $n$, $p$.
  $$P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}$$.
- **Poisson:** Número de eventos em intervalo de tempo/espaço, com taxa $\lambda$.
  $$P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}$$. É limite da Binomial para $n \to \infty$ e $p \to 0$.
- **Normal (Gaussiana):** Distribuição contínua, caracterizada por média $\mu$ e variância $\sigma^2$:
  $$
  f(x) = \frac{1}{\sqrt{2 \pi \sigma^2}} e^{ -\frac{(x-\mu)^2}{2\sigma^2} }
  $$

### 4. Esperança e Variância
- **Esperança (Valor Esperado):**
  $$E[X] = \sum x_i P(X=x_i)$$ para variável discreta, ou
  $$E[X] = \int x f(x) dx$$ para contínua.
- **Variância:**
  $$\text{Var}(X) = E[(X - E[X])^2] = E[X^2] - (E[X])^2$$.


A função acumulada, ou função distribuição acumulada (FAC), em estatística e probabilidade representa a soma acumulada dos valores ou a probabilidade acumulada até um certo ponto da variável.

### 5. Função Acumulada
Para uma sequência de dados, a função acumulada é a soma acumulativa dos valores até cada índice. Em Python, pode ser calculada com a função `np.cumsum`.

### Propriedades da Função Acumulada
- **Monotonicidade crescente:** A função acumulada nunca diminui, pois soma valores positivos.
- **Limites:** Começa em zero (ou no primeiro valor acumulado) e termina no total da soma dos dados.
- **Contínua à direita:** Na distribuição acumulada de uma variável aleatória, a função é contínua pela direita.
- **Escopo:** Pode ser aplicada para variáveis discretas (somatório) ou contínuas (integral acumulada).

### 6. Aplicações em Python

```python
#Exemplo:
import numpy as np

dados = np.array([1, 2, 3, 4, 5])
funcao_acumulada = np.cumsum(dados)
print(funcao_acumulada)  # Saída: [ 1, 3, 6, 10, 15]

#Aqui, o valor acumulado em cada posição é a soma dos elementos anteriores mais o atual [1, 1+2=3, 3+3=6, 6+4=10, 10+5=15].
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom, poisson, norm

# Bernoulli p=0.3
p = 0.3
x_bern = [0, 1]
pmf_bern = bernoulli.pmf(x_bern, p)
print("Bernoulli PMF:", list(zip(x_bern, pmf_bern)))

# Binomial n=10, p=0.5
n, p_bin = 100, 0.8
x_bin = np.arange(0, n+1)
pmf_bin = binom.pmf(x_bin, n, p_bin)
plt.figure(figsize=(10, 6))
plt.bar(x_bin, pmf_bin, alpha=0.7)
plt.title('Distribuição Binomial (n=350, p=0.8)')
plt.xlabel('k (número de sucessos)')
plt.ylabel('P(X=k)')
plt.show()

# Poisson λ=3
lambda_poisson = 15
x_pois = np.arange(0, 20)
pmf_pois = poisson.pmf(x_pois, lambda_poisson)
plt.figure(figsize=(10, 6))
plt.bar(x_pois, pmf_pois, alpha=0.75, color='green')
plt.title('Distribuição Poisson (λ=5)')
plt.xlabel('k (número de eventos)')
plt.ylabel('P(X=k)')
plt.show()

# Normal μ=0, σ=1
mu, sigma = 2, 3
x_norm = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
pdf_norm = norm.pdf(x_norm, mu, sigma)
plt.figure(figsize=(10, 6))
plt.plot(x_norm, pdf_norm, color='red')
plt.title('Distribuição Normal (μ=2, σ=3)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.show()

# Esperança e Variância exemplos
esperanca_bin = n * p_bin
variancia_bin = n * p_bin * (1 - p_bin)
print(f"Esperança Binomial: {esperanca_bin}")
print(f"Variância Binomial: {variancia_bin}")

esperanca_poisson = lambda_poisson
variancia_poisson = lambda_poisson
print(f"Esperança Poisson: {esperanca_poisson}")
print(f"Variância Poisson: {variancia_poisson}")
```
